## Data Collection

We collect the dataset from **Hugging Face** for **Khmer news classification**.  
The dataset consists of **6 distinct classes**, representing different news categories.

This dataset is used for training and evaluating a machine learning model for Khmer text classification.


```python
#!/usr/bin/env python3
"""
Script to download Khmer News Classification dataset
Save to: data/raw/khmer_news/
"""

import os
import sys
from pathlib import Path
from datasets import load_dataset
import pandas as pd
import pickle
from datetime import datetime

def create_directories():
    """Create necessary directories if they don't exist"""
    base_dir = Path(__file__).parent if '__file__' in globals() else Path.cwd()
    
    directories = [
        base_dir / 'data' / 'raw' / 'khmer_news',
        base_dir / 'data' / 'processed',
    ]
    
    for directory in directories:
        directory.mkdir(parents=True, exist_ok=True)
        print(f"✓ Directory ready: {directory}")
    
    return base_dir

def download_dataset(base_dir):
    """Download the Khmer News Classification dataset"""
    print("\n" + "="*60)
    print("Downloading Khmer News Classification Dataset")
    print("="*60)
    
    try:
        # Load dataset from Hugging Face
        print("\n[1/4] Loading dataset from Hugging Face...")
        dataset = load_dataset("CADT-IDRI/Khmer_News_classification")
        
        print(f"✓ Dataset loaded successfully!")
        print(f"✓ Available splits: {list(dataset.keys())}")
        
        # Save raw data
        print(f"\n[2/4] Saving raw data...")
        save_dir = base_dir / 'data' / 'raw' / 'khmer_news'
        
        saved_files = []
        
        # Save each split as CSV and pickle
        for split_name, split_data in dataset.items():
            df = pd.DataFrame(split_data)
            
            # Save as CSV
            csv_path = save_dir / f'{split_name}.csv'
            df.to_csv(csv_path, index=False, encoding='utf-8')
            saved_files.append(str(csv_path))
            print(f"  ✓ Saved {len(df):,} samples to {csv_path.name}")
            
            # Save as pickle for faster loading
            pkl_path = save_dir / f'{split_name}.pkl'
            df.to_pickle(pkl_path)
            saved_files.append(str(pkl_path))
            print(f"  ✓ Saved as pickle to {pkl_path.name}")
        
        # Save dataset metadata
        print(f"\n[3/4] Saving metadata...")
        metadata = {
            'name': 'Khmer_News_classification',
            'source': 'CADT-IDRI/Khmer_News_classification',
            'downloaded_at': datetime.now().isoformat(),
            'splits': list(dataset.keys()),
            'train_size': len(dataset['train']) if 'train' in dataset else 0,
            'test_size': len(dataset['test']) if 'test' in dataset else 0,
            'features': list(dataset['train'].features.keys()) if 'train' in dataset else [],
            'saved_files': saved_files
        }
        
        metadata_path = save_dir / 'metadata.pkl'
        with open(metadata_path, 'wb') as f:
            pickle.dump(metadata, f)
        print(f"  ✓ Metadata saved to {metadata_path.name}")
        
        # Display dataset information
        print(f"\n[4/4] Dataset Summary")
        print("="*60)
        print(f"Total splits: {len(dataset)}")
        
        for split_name, split_data in dataset.items():
            print(f"\n📊 {split_name.upper()} split:")
            print(f"   • Samples: {len(split_data):,}")
            print(f"   • Features: {', '.join(split_data.features.keys())}")
            
            # Show first sample
            if len(split_data) > 0:
                print(f"\n   Sample preview:")
                sample = split_data[0]
                for key, value in sample.items():
                    display_value = str(value)[:80] + "..." if len(str(value)) > 80 else str(value)
                    print(f"     • {key}: {display_value}")
        
        # Basic statistics
        print("\n" + "="*60)
        print("📈 Statistics")
        print("="*60)
        
        if 'train' in dataset:
            train_df = pd.DataFrame(dataset['train'])
            
            # Label distribution
            if 'label' in train_df.columns:
                print("\nLabel distribution:")
                label_counts = train_df['label'].value_counts().sort_index()
                for label, count in label_counts.items():
                    percentage = (count / len(train_df) * 100)
                    print(f"  Label {label}: {count:,} samples ({percentage:.2f}%)")
            
            # Text length statistics
            if 'text' in train_df.columns:
                train_df['text_length'] = train_df['text'].apply(len)
                print("\nText length statistics (characters):")
                stats = train_df['text_length'].describe()
                print(f"  • Mean: {stats['mean']:.0f}")
                print(f"  • Min: {stats['min']:.0f}")
                print(f"  • Max: {stats['max']:.0f}")
                print(f"  • Median: {stats['50%']:.0f}")
        
        print("\n" + "="*60)
        print("✅ Download completed successfully!")
        print("="*60)
        print(f"\n📁 Files saved to: {save_dir.absolute()}")
        print(f"\nSaved files:")
        for file in saved_files:
            print(f"  • {Path(file).name}")
        
        return True
        
    except Exception as e:
        print(f"\n❌ Error downloading dataset: {e}")
        print(f"\nError type: {type(e).__name__}")
        print("\nTroubleshooting tips:")
        print("  1. Check your internet connection")
        print("  2. Ensure you have the 'datasets' package installed:")
        print("     pip install datasets")
        print("  3. Try running with --verbose flag for more details")
        print("  4. Check if the dataset is still available at:")
        print("     https://huggingface.co/datasets/CADT-IDRI/Khmer_News_classification")
        
        return False

def main():
    """Main function"""
    print("\n🇰🇭 Khmer News Dataset Downloader")
    print(f"Started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    
    # Create directories
    base_dir = create_directories()
    
    # Download dataset
    success = download_dataset(base_dir)
    
    if success:
        print("\n✅ All done! You can now proceed with data preprocessing.")
        sys.exit(0)
    else:
        print("\n❌ Download failed. Please check the errors above.")
        sys.exit(1)

if __name__ == "__main__":
    main()
```
